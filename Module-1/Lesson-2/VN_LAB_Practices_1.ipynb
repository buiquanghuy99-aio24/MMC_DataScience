```python
#UCLN:
def greatest_common_divisor(a, b):
  common_divisor = []
  for i in range(1, min(a, b)):
    if a % i == 0 and b % i == 0:
      common_divisor.append(i)
  if common_divisor == []:
    return print(f'{a} and {b} have no common divisor')
  else:
    return print(f'{a} and {b} have common divisor is: {max(common_divisor)}')

greatest_common_divisor(0, 0)
greatest_common_divisor(0, 18)
greatest_common_divisor(174, 18)
```

```python
#UCLN: phân tích thừa số nguyên tố
def factorize(number):
  i = 2
  result = []
  while number != 1 and number > 1:
    if number % i == 0:
      number //= i
      result.append(i)
    else:
      i += 1
  return result

def greatest_common_divisor(a, b):
  factors_a = factorize(a)
  factors_b = factorize(b)
  divisors = factors_a + factors_b
  common_divisors = list(set([i for i in divisors if i in factors_a and i in factors_b]))
  greatest_common_divisor = 1
  for i in common_divisors:
    greatest_common_divisor *= i
  return greatest_common_divisor

greatest_common_divisor(0, 0)
greatest_common_divisor(0, 18)
greatest_common_divisor(174, 18)
```

Viết chương trình thực hiện nhập một xâu ký tự

a) Tìm từ dài nhất trong xâu đó.

b) Từ đó xuất hiện ở vị trí nào?
(Chú ý. nếu có nhiều từ có độ dài giống nhau thì chọn từ đầu tiên tìm thấy).

```python
def find_longest_word(text = input("Enter a text: ")):
  word_lst = text.split()
  longest_word = word_lst[0]
  for word in word_lst:
    if len(word) > len(longest_word):
      longest_word = word
  index = text.find(longest_word)

  return print(f'The longest word in text is: {longest_word} and it is at {index}')

find_longest_word()
```

## Question 2:

Viết chương trình tính n! với n là số tự nhiên không âm nhập vào từ bàn phím. Trong đó: 0! = 1 n! = 1.2.3...n

```python
def factorial(number = int(input("Enter a non-negative integer: "))):
  result = 1
  if number == 0:
    return result
  else:
    for i in range(1, number + 1):
      result *= i
  return result

factorial()
```
