## Câu hỏi 6: Liệt kê các trận đấu và thông tin về leage và country tương ứng

Bảng Match có quan hệ với :
 - Bảng Country thông qua khóa country_id
 - Bảng League thông qua khóa League_id

Sử dụng câu lệnh join để kết nối dữ liệu từ 3 bảng trên.


```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("""SELECT      c.name AS 'Country Name', l.name AS 'League Name', m.*
                        FROM        Match   AS m
                        LEFT JOIN   Country AS c    ON c.id = m.country_id
                        LEFT JOIN   League  AS l    ON l.id = m.league_id
                        """, conn)
result
```

## Câu hỏi 7: Liệt kê số lượng trận đấu của mỗi giải đấu bao gồm cả tên giải đấu, sắp xếp theo thứ tự giảm dần số trận đấu

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("""SELECT      m.league_id, l.name AS 'league name', COUNT(m.id) AS 'num match'
                        FROM        Match   AS m
                        LEFT JOIN   League  AS l    ON l.id = m.league_id
                        GROUP BY    m.league_id
                        ORDER BY    COUNT(m.id) DESC
                        """, conn)
result
```

## Câu hỏi 12: Liệt kê tên các đội bóng và tổng số bàn thắng mỗi đội, sắp xếp  theo số lượng giảm dần

- Tổng số bàn thắng của một đội bao gồm cả số bàn thắng ở sân nhà và sân khách.
- Sử dụng câu lệnh Union  

```python
# ĐIỀN CODE VÀO ĐÂY

home_away_goals = pd.read_sql("""SELECT     t.team_long_name, SUM(m.home_team_goal) AS 'Goals'
                        FROM        Match   AS m
                        LEFT JOIN   Team    AS t    ON t.team_api_id = m.home_team_api_id
                        GROUP BY    t.team_long_name
                        UNION ALL
                        SELECT     t.team_long_name, SUM(m.away_team_goal) AS 'Goals'
                        FROM        Match   AS m
                        LEFT JOIN   Team    AS t    ON t.team_api_id = m.away_team_api_id
                        GROUP BY    t.team_long_name
                        """, conn)
#Do mỗi đội bóng sẽ xuất hiện 2 lần khi Union (1 lần cho home_goal, 1 lần cho away goal) nên cần group bảng:
total_goals_by_team = home_away_goals.groupby('team_long_name')['Goals'].sum().reset_index()
total_goals_by_team
```

```python
top_20_teams = total_goals_by_team.nlargest(5, 'Goals')
top_20_teams
```

```python
# Solution 2
total_goals_by_team = pd.read_sql("""
                                    SELECT      t.team_long_name,
                                                SUM(
                                                    CASE WHEN m.home_team_api_id = t.team_api_id    THEN m.home_team_goal ELSE 0 END
                                                    + CASE WHEN m.away_team_api_id = t.team_api_id  THEN m.away_team_goal ELSE 0 END
                                                    ) AS 'Goals'
                                    FROM        Match   AS m
                                    LEFT JOIN   Team    AS t    ON  t.team_api_id = m.home_team_api_id
                                                                OR  t.team_api_id = m.away_team_api_id
                                    GROUP BY    t.team_long_name
                                """, conn)

total_goals_by_team
```

## Câu hỏi 13: Liệt kê tên các đội bóng (long name) và tổng số trận đấu đội đó tham gia, sắp xếp theo thứ tự giảm dần.

```python
home_away_matchs = pd.read_sql("""SELECT     t.team_long_name, COUNT(m.match_api_id) AS 'Matchs'
                        FROM        Match   AS m
                        LEFT JOIN   Team    AS t    ON t.team_api_id = m.home_team_api_id
                        GROUP BY    t.team_long_name
                        UNION ALL
                        SELECT     t.team_long_name, COUNT(m.match_api_id) AS 'Matchs'
                        FROM        Match   AS m
                        LEFT JOIN   Team    AS t    ON t.team_api_id = m.away_team_api_id
                        GROUP BY    t.team_long_name
                        """, conn)

total_matchs_by_team = home_away_matchs.groupby('team_long_name')['Matchs'].sum().reset_index()
total_matchs_by_team
```

```python
# ĐIỀN CODE VÀO ĐÂY
total_matchs_by_team = pd.read_sql("""
                                    SELECT      t.team_long_name,
                                                COUNT(
                                                    CASE WHEN m.home_team_api_id = t.team_api_id    THEN m.match_api_id ELSE 0 END
                                                    + CASE WHEN m.away_team_api_id = t.team_api_id  THEN m.match_api_id ELSE 0 END
                                                    ) AS 'Matchs'
                                    FROM        Match   AS m
                                    LEFT JOIN   Team    AS t    ON  t.team_api_id = m.home_team_api_id
                                                                OR  t.team_api_id = m.away_team_api_id
                                    GROUP BY    t.team_long_name
                                """, conn)

total_matchs_by_team
```

## Câu hỏi 14: Liệt kê số trận thắng, thua và hòa của mỗi đội.

- Nếu một đội có số lượng bàn thắng trên sân nhà > số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thắng" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà < số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thua" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà = số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "hòa" trong trận đấu này
- Sử dụng câu lệnh "Case When"


```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("""SELECT      t.team_long_name,
                                    COUNT(CASE
                                            WHEN m.home_team_api_id = t.team_api_id AND m.home_team_goal > m.away_team_goal THEN 1
                                            WHEN m.away_team_api_id = t.team_api_id AND m.home_team_goal < m.away_team_goal THEN 1
                                        END) As wins,
                                    COUNT(CASE
                                            WHEN m.home_team_api_id = t.team_api_id AND m.home_team_goal < m.away_team_goal THEN 1
                                            WHEN m.away_team_api_id = t.team_api_id AND m.home_team_goal > m.away_team_goal THEN 1
                                        END) As losses,
                                    COUNT(CASE
                                            WHEN m.home_team_api_id = t.team_api_id AND m.home_team_goal = m.away_team_goal THEN 1
                                            WHEN m.away_team_api_id = t.team_api_id AND m.home_team_goal = m.away_team_goal THEN 1
                                        END) As draws                                 
                        FROM        Match   AS m
                        LEFT JOIN   Team    As t    ON t.team_api_id = m.home_team_api_id
                                                    OR t.team_api_id = m.away_team_api_id
                        GROUP BY    t.team_long_name
                        ORDER BY    wins DESC
                        """, conn)
result
```
